# CNN Model Architectures

```python
def CNN_Model_1():
    
    # Set Device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Build the model.
    model = nn.Sequential()
    
    model.add_module('conv1', nn.Conv2d(in_channels=3, out_channels = 16, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv1', nn.ReLU())
    model.add_module('pool1', nn.MaxPool2d(2,2))
    
    model.add_module('conv2', nn.Conv2d(in_channels=16, out_channels = 32, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv2', nn.ReLU())
    model.add_module('pool2', nn.MaxPool2d(2,2))
    
    model.add_module('conv3', nn.Conv2d(in_channels=32, out_channels = 64, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv3', nn.ReLU())
    model.add_module('pool3', nn.MaxPool2d(2,2))
    
    model.add_module('conv4', nn.Conv2d(in_channels=64, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv4', nn.ReLU())
    model.add_module('pool4', nn.MaxPool2d(2,2))
    
    model.add_module("flat1", nn.Flatten())
    
    model.add_module('fulc1', nn.Linear(14*14*128, 128))
    model.add_module('actv5', nn.ReLU())
    
    model.add_module('fulc3', nn.Linear(128, 4))
    model.add_module('soft1', nn.Softmax(dim=1))

    return model
```

```python
def CNN_Model_2():
    
    # Set Device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Build the model.
    model = nn.Sequential()
    
    model.add_module('conv1', nn.Conv2d(in_channels=3, out_channels = 32, kernel_size = 5, padding = 2, stride = 1))
    model.add_module('actv1', nn.ReLU())
    model.add_module('pool1', nn.MaxPool2d(2,2))
    
    model.add_module('conv2', nn.Conv2d(in_channels=32, out_channels = 64, kernel_size = 5, padding = 2, stride = 1))
    model.add_module('actv2', nn.ReLU())
    model.add_module('pool2', nn.MaxPool2d(2,2))
    
    model.add_module('conv3', nn.Conv2d(in_channels=64, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv3', nn.ReLU())
    model.add_module('pool3', nn.MaxPool2d(2,2))
    
    model.add_module('conv4', nn.Conv2d(in_channels=128, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv4', nn.ReLU())
    model.add_module('pool4', nn.MaxPool2d(2,2))
    
    model.add_module("flat1", nn.Flatten())
    
    model.add_module('fulc1', nn.Linear(14*14*128, 512))
    model.add_module('actv5', nn.ReLU())
    
    model.add_module('fulc2', nn.Linear(512, 4))
    model.add_module('soft1', nn.Softmax(dim=1))

    return model
```

```python
def CNN_Model_3():
    
    # Set Device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Build the model.
    model = nn.Sequential()
    
    model.add_module('conv1', nn.Conv2d(in_channels=3, out_channels = 32, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv1', nn.ReLU())
    model.add_module('pool1', nn.MaxPool2d(2,2))
    model.add_module('batc1', nn.BatchNorm2d(32))
    
    model.add_module('conv2', nn.Conv2d(in_channels=32, out_channels = 64, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv2', nn.ReLU())
    model.add_module('pool2', nn.MaxPool2d(2,2))
    model.add_module('batc2', nn.BatchNorm2d(64))
    
    model.add_module('conv3', nn.Conv2d(in_channels=64, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv3', nn.ReLU())
    model.add_module('pool3', nn.MaxPool2d(2,2))
    model.add_module('batc3', nn.BatchNorm2d(128))
    
    model.add_module('conv4', nn.Conv2d(in_channels=128, out_channels = 256, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv4', nn.ReLU())
    model.add_module('pool4', nn.MaxPool2d(2,2))
    model.add_module('batc4', nn.BatchNorm2d(256))

    model.add_module("flat1", nn.Flatten())
    
    model.add_module('fulc1', nn.Linear(14*14*256, 512))
    model.add_module('actv5', nn.ReLU())
    model.add_module('drop1', nn.Dropout(0.5))
    
    model.add_module('fulc2', nn.Linear(512, 256))
    model.add_module('actv5', nn.ReLU())
    model.add_module('drop2', nn.Dropout(0.5))
    
    model.add_module('fulc3', nn.Linear(256, 4))
    model.add_module('soft1', nn.Softmax(dim=1))
    
    return model
```

```python
def CNN_Model_4():
    
    # Set Device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Build the model.
    model = nn.Sequential()

    model.add_module('conv1', nn.Conv2d(in_channels=3, out_channels = 16, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv1', nn.ReLU())
    model.add_module('conv2', nn.Conv2d(in_channels=16, out_channels = 32, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv2', nn.ReLU())
    model.add_module('pool1', nn.MaxPool2d(2,2))

    model.add_module('conv3', nn.Conv2d(in_channels=32, out_channels = 64, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv3', nn.ReLU())
    model.add_module('conv4', nn.Conv2d(in_channels=64, out_channels = 64, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv4', nn.ReLU())
    model.add_module('pool2', nn.MaxPool2d(2,2))

    model.add_module('conv5', nn.Conv2d(in_channels=64, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv5', nn.ReLU())
    model.add_module('conv6', nn.Conv2d(in_channels=128, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv6', nn.ReLU())
    model.add_module('pool3', nn.MaxPool2d(2,2))
    
    model.add_module('conv7', nn.Conv2d(in_channels=128, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv7', nn.ReLU())
    model.add_module('pool4', nn.MaxPool2d(2,2))

    model.add_module("flat1", nn.Flatten())

    model.add_module('fulc1', nn.Linear(14*14*128, 512))
    model.add_module('actv14', nn.ReLU())
    model.add_module('drop1', nn.Dropout(0.5))

    model.add_module('fulc2', nn.Linear(512, 128))
    model.add_module('actv15', nn.ReLU())
    model.add_module('drop2', nn.Dropout(0.5))

    model.add_module('fulc3', nn.Linear(128, 4))               
    model.add_module('soft1', nn.Softmax(dim=1))

    return model
```

```python
def CNN_Model_5():
    
    # Set Device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Build the model.
    model = nn.Sequential()

    model.add_module('conv1', nn.Conv2d(in_channels=3, out_channels = 16, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv1', nn.ReLU())
    model.add_module('conv2', nn.Conv2d(in_channels=16, out_channels = 32, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv2', nn.ReLU())
    model.add_module('pool1', nn.MaxPool2d(2,2))
    model.add_module('batc1', nn.BatchNorm2d(32))

    model.add_module('conv3', nn.Conv2d(in_channels=32, out_channels = 64, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv3', nn.ReLU())
    model.add_module('conv4', nn.Conv2d(in_channels=64, out_channels = 64, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv4', nn.ReLU())
    model.add_module('pool2', nn.MaxPool2d(2,2))
    model.add_module('batc2', nn.BatchNorm2d(64))

    model.add_module('conv5', nn.Conv2d(in_channels=64, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv5', nn.ReLU())
    model.add_module('conv6', nn.Conv2d(in_channels=128, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv6', nn.ReLU())
    model.add_module('pool3', nn.MaxPool2d(2,2))
    model.add_module('batc3', nn.BatchNorm2d(128))
    
    model.add_module('conv7', nn.Conv2d(in_channels=128, out_channels = 128, kernel_size = 3, padding = 1, stride = 1))
    model.add_module('actv7', nn.ReLU())
    model.add_module('pool4', nn.MaxPool2d(2,2))
    model.add_module('batc4', nn.BatchNorm2d(128))

    model.add_module("flat1", nn.Flatten())

    model.add_module('fulc1', nn.Linear(14*14*128, 1024))
    model.add_module('actv14', nn.ReLU())
    model.add_module('drop1', nn.Dropout(0.5))

    model.add_module('fulc2', nn.Linear(1024, 512))
    model.add_module('actv15', nn.ReLU())
    model.add_module('drop2', nn.Dropout(0.5))

    model.add_module('fulc3', nn.Linear(512, 4))               
    model.add_module('soft1', nn.Softmax(dim=1))

    return model
```
